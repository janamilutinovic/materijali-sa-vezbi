# Furijeove transformacije &ndash; uvodni primer

Funkcija $f(t)$ čiji je argument vreme $t$ se naziva **signal**. Signal se, sa jedne strane, može predstaviti svojom jačinom u zavisnosti od vremena. Ovakvo predstavljanje se vrši u **vremenskom domenu** (eng. *time domain*). Sa druge strane, nekad je zgodno analizirati koliko je svaka frekvencija zastupljena u tom signalu, pa tada govorimo o predstavljanju signala u **frekvencijskom domenu** (eng. *frequency domain*). Oba prikaza su samo dva načina predstavljanja iste funkcije. **Furijeova transformacija** omogućava prelazak iz vremenskog u frekvencijski domen, a **inverzna Furijeova transormacija** prelazak iz frekvencijskog u vremenski domen.

Na gornjoj animaciji je najpre crvenom bojom u vremenskom domenu prikazana polazna funkcija $f$. Plavom bojom je prikazano njeno razlaganje na linearnu kombinaciju sinusa i kosinusa. Dalje se može videti kako se svaka od ovih sinusnih i kosinusnih komponenti širi u frekvencijskom domenu. Oznaka $\hat{f}$ se odnosi na predstavljanje ove funkcije u frekvencijskom domenu.

Posmatrajmo, na primer, funkciju $f(t) = \sin t + 0.1 \sin 3t$, koja je na donjoj slici prikazana crvenom bojom. Ona se može razložiti na dve sinusoide, koje su na slici predstavljene zelenom i plavom bojom. Drugim rečima, u svakom trenutnku saberemo vrednosti vremenske koordinate (na $x$-osi) plavog i zelenog grafika, dobijamo vrednost vremenske koordinate crvenog grafika. Možemo videti da je zelena sinusoida periodična na grafiku sa periodom nešto većim od 6, preciznije $2\pi$. To znači da bi ona mogla predstavljati tačku na krugu koja obiđe ceo krug za jednu sekundu, neka to predstavlja frekvenciju $\omega = 1$. Slično, plava sinusoida može predstavljati tačku na krugu koja tri puta obiđe krug za jednu sekundu, pa to može biti predstavljeno frekvencijom $\omega = 3$. Vidimo da zelena sinusoida ima znatno veći uticaj na rezultujuću krivu, preciznije kažemo da je $\hat{f}(1) = 1$. Plava ima 10 puta manji, i važi $\hat{f}(3) = 0.1$. Ostale frekvencije za ovu funkciju nisu uopšte ni prisutne, pa je u tim slučajevima $\hat{f}(\omega) = 0$.

Posmatrajmo proizvoljnu sinusoidu $a \sin(2\pi f_0 t + \phi)$. Ako uvedemo oznaku $\omega_0 = 2\pi f_0$, sinusoida se može napisati i u obliku $a \sin(\omega_0 t + \phi)$. Parametar $a$ se naziva **amplituda** &ndash; na grafiku amplituda zelene sinusoide je 1, a plave 0.1. Parametar $f_0$ predstavlja **osnovnu frekvenciju**, odnosno broj obrtaja koji neka tačka, krećući se po krugu, može obići za jednu sekundu, a $\omega$ **ugaonu frekvenciju**. Parametar $\phi$ predstavlja **fazu**, odnosno meru koliko je sinusoida pomerena na $x$-osi u odnosu na koordinatni početak. Primetimo i da je grafik kosinusne funkcije takođe sinusoida, pomerena po $x$-osi za vrednost $\pi / 2$. U literaturi se često oznake poput $2f_0$, $3f_0$, odnosno $nf_0$, nazivaju drugim, trećim, odnosno **harmonicima** višeg reda.

Furijeova transformacija funkcije $f$ se najčešće označava sa $\hat{f}$. Pretpostaka da je funkcija $f: \mathbb{R} \rightarrow \mathbb{C}$ apsolutno integrabilna. Furijeova transformacija je funkcija $$\hat{f}(u) = \int_{-\infty}^{\infty} f(t) e^{2\pi i u t}dt.$$ Nasuprot njoj, definiše se inverzna Furijeova transformacija sa $$f(t) = \int_{-\infty}^{\infty} \hat{f}(u) e^{-2\pi i u t}dt.$$ Ovde je sa $t$ predstavljeno vreme, a sa $u$ frekvencija. 

Međutim, u praksi se najčešće govori o **diskretnoj Furijeovoj transformaciji**, budući da je funkcija poznata samo na konačnom skupu tačaka. Tada se, prirodno, integral zamenjuje odgovarajućom sumom. Neka niz $f_0, f_1, \dots, f_{N-1}$, predstavlja niz od $N$ kompleksnih brojeva, koji može predstavljati vrednosti funkcije $f$ u nekih $N$ tačaka. Diskretna Furijeova transformacija transformiše polazni niz u novi niz kompleksnih brojeva $\hat{f}_0, \hat{f}_1, \dots, \hat{f}_{N-1}$ i data je $$\hat{f}_k = \sum_{n=0}^{N-1} f_n e^{-2\pi ikn/N}.$$ Imajući u vidu Ojlerov identitet $e^{i\phi} = \cos \phi + i \sin \phi$, prethodni izraz se može napisati u obliku $$\hat{f}_k = \sum_{n=0}^{N-1} f_n \left( \cos \frac{2\pi kn}{N} - i \sin \frac{2\pi kn}{N} \right).$$

Nasuprot tome, može se definisati **inverzna diskretna Furijeova** transformacija sa $$f_n=\frac{1}{N} \sum_{k=0}^{N-1} \hat{f}_k e^{2\pi i kn/N}.$$

Ako se na svakih $T_s$ sekundi vrši merenje signala, $T_s$ se naziva **period uzrokovanja**. **Frekvencija uzrokavanja** $F_s$ je obrnuto srazmerna periodu uzrokovanja i izražena je u hercima. Važi da je $F_s = 1 / T_s$. Na primer, ako je $T_s = 0.5$ s, tada je $F_s = 1 / 0.5 = 2$ Hz, odnosno signal se meri dva puta u sekundi. Ako je ukupan broj uzoraka jednak $N$, tada se $T = NT_s$ naziva **period uzrokovanja**. **Frekvencijka rezolucija** je vrednost $\Delta f = F_s / N$, a $k$-ti **frekvencijski podeok** vrednost $k \Delta f$.

**Primer.** Posmatrajmo signal $f(t) = 2\sin (2\pi t) + 3 \sin (44\pi t) + \epsilon$, gde je $\epsilon$ šum, koji ćemo generisati proizvoljno iz standardne normalne raspodele. Neka je signal izmeren 6000 puta tokom 100 sekundi. 

Iz zadatka sledi da je $N = 6000$ i $T = 100$. Odatle je $T_s = T / N = 1 / 60$ s i $F_s = 1 / T = 60$ Hz, što znači da se signal meri 60 puta u sekundi. Najpre ćemo nacrtati signal u vremenskom domenu. Definisaćemo funkciju `test_signal` koja vraća vrednost $f$ i nju testirati na 6000 ekvidistantnih tačaka u intervalu $[0, 100]$

```python
import numpy as np
import matplotlib.pyplot as plt
# paket za Furijeove transformacije
from numpy import fft
#from scipy import signal
# za inline prikazivanje matplotlib grafika
%matplotlib inline

def test_signal(t):
    n = t.shape[0]
    noise = np.random.randn(n)
    return 2*np.sin(2*np.pi*t) + 3*np.sin(44*np.pi*t) + noise

t = np.linspace(0, 100, 6000, endpoint=True)
f = test_signal(t)
plt.plot(t, f)
plt.show()
```

Da bismo jasnije videli kako izgleda signal u vremenskom domenu, koncentrisaćemo se, na primer, na segment $[0, 5]$:

```python
plt.plot(t, f)
plt.xlim(0, 5)
plt.show()
```

Izvršimo sada diskretnu Furijeovu transformaciju. Kao rezultat, za 6000 ulaznih brojeva, dobićemo novih 6000 kompleksnih brojeva.

```python
c = fft.fft(f)
print('Tip koeficijenata:', c.dtype)
print('Broj koeficijenata:', c.shape)
print('Prvih deset koeficijenata:\n', c[:10])
```

Generišimo sada niz frekvencija funkcijom `fftfreq`. Njeni argumenti su $N$ i $T_s$ (drugi argument se često označava i sa $1 / F_s$). Generisaće se redom vrednosti ekvidistantnih frekvencija, najpre u intervalu $[0, 30)$, a zatim u intervalu $[-30, 0)$. Niz frekvencija predstavlja $x$-osu u frekvencijskom domenu.

```python
freq = fft.fftfreq(6000, 1.0/60)
plt.plot(freq)
plt.show()
```

Prikaz funkcije $f$ u frekvencijskom domenu izgleda ovako:

```python
plt.figure(figsize=(18,5))
plt.xticks(np.linspace(-30,30,31))
# potrebno je uzeti module kompleksnih brojeva, sto radi funkcija abs
plt.plot(freq, np.abs(c))
plt.show()
```

Možemo videti da su ubedljivo najveće vrednosti frekvencija u tačkama $1$, $-1$, $22$ i $-22$, što je i očekivano, budući da, osim šuma, u funkciji figurišu $\sin (1 \cdot 2\pi t)$ i $\sin (22 \cdot 2\pi t)$. Šum utiče na znatno manje frekvencije na grafiku, koje se nalaze izmežu najvećih stubića u pomenutim tačkama. Inače, frekvencijski domen se najčešće predstavlja u obliku *stem* grafika:

```python
plt.stem(freq[::100], np.abs(c)[::100])
plt.show()
```

Odbacićemo šum, a zatim primeniti inverznu diskretnu Furijeovu transformaciju i funkciju ponovo prikazati u vremenskom domenu, uporedivši je sa polaznom:

```python
# odbacivanje suma - uzimaju se vrednosti vece od 1000
mask = np.abs(c) > 1000
c_filtered = c * mask
f_filtered = fft.ifft(c_filtered)
```

```python
plt.xlim(0, 5)
plt.plot(t, f, 'b-')
# nama je ovde bitan samo realan deo funkcije, buduci da je i polazna funkcija realna
plt.plot(t, f_filtered.real, 'r-')
plt.show()
```
